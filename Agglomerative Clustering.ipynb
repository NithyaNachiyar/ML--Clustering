```python
import pandas as pd
```

```python
dataset=pd.read_csv('Mall_Customers.csv')
```

```python
dataset
```

```python
X=dataset.iloc[:,3:5].values
```

```python
X
```

```python
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
dendrogram=sch.dendrogram(sch.linkage(X, method='ward'))
import matplotlib.pyplot as plt
plt.title('dendrogram')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
clustering=AgglomerativeClustering(n_clusters=5)
label=clustering.fit_predict(X)
```

```python
supervised=pd.DataFrame(dataset)
```

```python
supervised['Clustering Group']=label
```

```python
supervised
```

```python
import seaborn as sns
facet=sns.lmplot(supervised, x=supervised.columns[3], y=supervised.columns[4],hue=supervised.columns[5], 
                 fit_reg=True,  legend=True, legend_out=True)
```
