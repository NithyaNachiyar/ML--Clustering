```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv('Mall_Customers.csv')
```

```python
dataset
```

```python
x=dataset.iloc[:, 3:5].values
```

```python
x
```

```python
from sklearn.cluster import AffinityPropagation
clustering=AffinityPropagation(affinity='euclidean', random_state=0)
label=clustering.fit_predict(x)
```

```python
label
```

```python
dataset
```

```python
dataset['clustering Group']=label
```

```python
dataset
```

```python
import seaborn as sns
facet=sns.lmplot(dataset, x=dataset.columns[3], y=dataset.columns[4], hue=dataset.columns[5], fit_reg=True, legend=True, legend_out=True)
```
