```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv('Mall_Customers.csv')
```

```python
X=dataset.iloc[:, 3:5].values
```

```python
from sklearn.mixture import GaussianMixture
clustering=GaussianMixture(n_components=10,covariance_type='full', 
                           means_init=None, precisions_init=None, random_state=None, warm_start=False, 
                           verbose_interval=10)
label=clustering.fit_predict(X)
```

```python
label
```

```python
dataset['Clustering Group']=label
```

```python
supervised=pd.DataFrame(dataset)
```

```python
import seaborn as sns
facet=sns.lmplot(supervised, x=supervised.columns[3],y=supervised.columns[4], hue=supervised.columns[5], 
                 legend=True, legend_out=True,  fit_reg=False)
```
