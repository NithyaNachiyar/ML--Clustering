```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv('Mall_Customers.csv')
```

```python
dataset
```

```python
X=dataset.iloc[:, 3:5].values
```

```python
X
```

```python
from sklearn.cluster import KMeans
list=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i, random_state=40, init="k-means++")
    kmeans.fit(X)
    list.append(kmeans.inertia_)
plt.plot(range(1,11), list)
plt.title('The Elbow Method')
plt.xlabel('No. of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
from sklearn.cluster import KMeans
kmeans=KMeans(n_clusters=6, random_state=40, init="k-means++")
y_mean=kmeans.fit_predict(X)
```

```python
y_mean
```

```python
supervised=pd.DataFrame(dataset)
```

```python
supervised['Clustering Group']=y_mean
```

```python
supervised
```
