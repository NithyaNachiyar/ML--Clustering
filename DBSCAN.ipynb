```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("Mall_Customers.csv")
```

```python
X=dataset.iloc[:, 3:5].values
```

```python
X
```

```python
from sklearn.cluster import DBSCAN
clustering=DBSCAN(eps=3,min_samples=3)
label=clustering.fit_predict(X)
```

```python
label
```

```python
dataset['Clustering Group']=label
```

```python
dataset
```

```python
supervised=pd.DataFrame(dataset)
```

```python
import seaborn as sns
facet=sns.lmplot(supervised, x=supervised.columns[3], y=supervised.columns[4], hue=supervised.columns[5], fit_reg=True, legend=True, legend_out=True)
```
