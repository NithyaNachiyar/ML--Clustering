```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv('Mall_Customers.csv')
dataset
```

```python
X=dataset.iloc[:, 3:5].values
X
```

```python
from sklearn.cluster import MeanShift
clustering=MeanShift(bandwidth=5)
label=clustering.fit_predict(X)
```

```python
label
```

```python
dataset['clustering Group']=label
```

```python
dataset
```

```python
import seaborn as sns
facet=sns.lmplot(dataset, x=dataset.columns[3], y=dataset.columns[4], hue=dataset.columns[5], 
                 legend=True, legend_out=True, fit_reg=True)
```
